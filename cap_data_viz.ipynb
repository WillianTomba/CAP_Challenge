```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar o arquivo CSV
file_path = 'goals_by_team.csv'
df = pd.read_csv(file_path)

# Classificar os dados pelo número de gols em ordem decrescente
df = df.sort_values(by='Goals', ascending=False)

# Gerar o gráfico de barras
plt.figure(figsize=(12, 8))
bars = plt.bar(df['Team'], df['Goals'], color='skyblue')

# Adicionar rótulos aos dados
for bar in bars:
    yval = bar.get_height()
    # Ajustar a posição dos rótulos para ficarem dentro dos limites do gráfico
    if yval == bars[0].get_height() or yval == bars[-1].get_height():
        plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.2, int(yval), ha='center', va='bottom')
    else:
        plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.5, int(yval), ha='center', va='bottom')

plt.xlabel('Seleção')
plt.ylabel('Gols (descontando os pênaltis)')
plt.title('Gols por Seleção nas Eliminatórias da Copa do Mundo Feminina 2023')
plt.xticks(rotation=90)
plt.tight_layout()

# Salvar o gráfico como uma imagem
plt.savefig('goals_by_team_sorted_plot.png')

# Exibir o gráfico
plt.show()
```
